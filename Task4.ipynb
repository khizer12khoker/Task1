```python
from mistralai import Mistral

API_KEY = "gr64pOPqJniL9d4WVmiaUPVMIvYzylyE"

client = Mistral(api_key=API_KEY)
```

```python
def safety_filter(text):
    blocked_words = [
        "dosage", "dose", "mg", "prescribe",
        "treatment", "cure"
    ]
    return not any(word in text.lower() for word in blocked_words)


def health_chatbot(user_input):

    if not safety_filter(user_input):
        return "‚ö†Ô∏è I can only provide general health information. Please consult a doctor."

    messages = [
        {
            "role": "system",
            "content": (
                "You are a friendly and responsible medical assistant.\n"
                "Rules:\n"
                "- Provide general health information only\n"
                "- Do NOT prescribe medicine or dosage\n"
                "- Keep answers simple and safe\n"
                "- Recommend seeing a doctor when needed"
            )
        },
        {
            "role": "user",
            "content": user_input
        }
    ]

    response = client.chat.complete(
        model="mistral-small-latest",
        messages=messages
    )

    return response.choices[0].message.content
```

```python
print("ü©∫ General Health Chatbot (type 'exit' to quit)\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Bot: Stay healthy! üëã")
        break

    reply = health_chatbot(user_input)
    print("Bot:", reply, "\n")
```
